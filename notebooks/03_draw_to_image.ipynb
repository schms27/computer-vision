## Create blank canvas

```python
canvas_b = np.zeros((1024,1024,3), np.uint8)
canvas_w = np.zeros((1024,1024,3), np.uint8)
canvas_w.fill(255)
show_image(canvas_b, "Black Canvas - RGB")
show_image(canvas_w, "White Canvas - RGB")
```

## Lines
Draw some single lines

```python
cv2.line(canvas_b, (0,1023), (1023,0), (255,127,0), 5)
cv2.line(canvas_b, (0,850), (355,0), (100,95,200), 8)
show_image(canvas_b,"Black canvas with lines")
```

## Rectangles and Circles
Draw some rectangles

```python
cv2.rectangle(canvas_w, (150,150), (350,600), (100,255,0), -1)
cv2.rectangle(canvas_w, (750,30), (30,850), (150,70,110), 8)

cv2.circle(canvas_w, (780, 950), 200, (15,150,50), 5) 
cv2.circle(canvas_w, (512, 512), 410, (125,125,125), 12) 
show_image(canvas_w,"White canvas with rectangles and circles")
```

## Polylines
Draw polyline

```python
canvas_b = np.zeros((1024,1024,3), np.uint8)

# define points of the polyline
points = np.array( [[50,50], [500,70], [800,100], [50,550], [980,1000]], np.int32)
print(f'Before reshaping: {points.shape}')

# reshape points in form  required by polylines
points = points.reshape((-1,1,2))
print(f'After reshaping: {points.shape}')

cv2.polylines(canvas_b, [points], True, (100,50,180), 3)
show_image(canvas_b, "Black Canvas with polygon")
```

## Read image and write some text to it

```python
image = cv2.imread('./tmp/wiki/01/1203301_1908-05-29_1958.jpg')
text = 'Is this real?'
cv2.putText(image, text, (50,150), cv2.FONT_HERSHEY_PLAIN, 3, (0,200,250), 4)

show_image(image, "Chair with text")
```
