## Loading images
Now we read some sample image using opencv and display it with matplotlib

```python
image = cv2.imread('./tmp/wiki/00/11328300_1980-06-10_2008.jpg')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## Saving Images
We can also use opencv to save images

```python
!mkdir ./out/
cv2.imwrite('./out/sample_output.jpg', image)
cv2.imwrite('./out/sample_output.png', image)
```
