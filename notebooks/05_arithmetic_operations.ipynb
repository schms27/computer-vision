## Change brightness

```python
# load an image as grayscale
img = cv2.imread('tmp/wiki/05/179105_1921-02-08_1943.jpg', 0)
show_image(img, "Grayscale image")

# Create a matrix of ones, then multiply it by a scaler of 100 
# This gives a matrix with same dimesions of our image with all values being 100
M = np.ones(img.shape, dtype = "uint8") * 100 

# we can use cv2.add to increase the brightness.
# Note: this works better than simply adding img + M, because we would experience clipping this way
added = cv2.add(img, M)
show_image(added, "Increased Brightness")

sub = cv2.subtract(img, M)
show_image(sub, "Decreased Brightness")
```

## Bitwise Operations (AND, OR, XOR, NOT)
This shows how to apply bitwise logical operations to images.


```python
# Making a square
square = np.zeros((400, 400), np.uint8)
cv2.rectangle(square, (100, 100), (300, 300), 255, -2)
show_image(square, "square")

# Making a ellipse
ellipse = np.zeros((400, 400), np.uint8)
cv2.ellipse(ellipse, (150, 200), (150, 100), 30, 0, 180, 255, -1)
show_image(ellipse, "ellipse")
```

```python
# intersection
img_and = cv2.bitwise_and(square, ellipse)
show_image(img_and, "AND")

# Shows where either square or ellipse is 
img_or = cv2.bitwise_or(square, ellipse)
show_image(img_or, "bitwiseOr")

# Shows where either exist by itself
img_xor = cv2.bitwise_xor(square, ellipse)
show_image(img_xor, "bitwiseXor")

# Shows everything that isn't part of the square
img_not = cv2.bitwise_not(square)
show_image(img_not, "bitwiseNot_sq")
```
